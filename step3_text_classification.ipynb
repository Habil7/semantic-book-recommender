```python
# Importing numpy for numerical operations
import numpy as np

# Importing pandas for data handling and data frames
import pandas as pd

# Importing seaborn for heatmap plotting
import seaborn as sns

# Importing matplotlib for plotting and visualization
import matplotlib.pyplot as plt

# Reading the cleaned books CSV file from the NLP Project directory in Google Drive
books = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/NLP/NLP Project/books_cleaned.csv")
```

```python
# Defining a mapping dictionary to group detailed categories into broader, simplified categories
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children's Fiction",
    "Biography & Autobiography": "Nonfiction",
    "History": "Nonfiction",
    "Literary Criticism": "Nonfiction",
    "Philosophy": "Nonfiction",
    "Religion": "Nonfiction",
    "Comics & Graphic Novels": "Fiction",
    "Drama": "Fiction",
    "Juvenile Nonfiction": "Children's Nonfiction",
    "Science": "Nonfiction",
    "Poetry": "Fiction"
}

# Creating a new column with simplified category labels based on the mapping dictionary
books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
# Importing the zero-shot classification pipeline from Hugging Face Transformers
from transformers import pipeline

# Defining the high-level categories that we want the model to classify into
fiction_categories = ["Fiction", "Nonfiction"]

# Creating a zero-shot classification pipeline
# Using the 'facebook/bart-large-mnli' model for natural-language inference
# Setting device = 0 to enable GPU acceleration if available
pipe = pipeline(
    "zero-shot-classification",
    model = "facebook/bart-large-mnli",
    device = 0
)
```

```python
# Selecting the first book description from the rows labeled as 'Fiction' in the simplified categories
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop = True)[0]
```

```python
# Running the zero-shot classifier on the selected description to predict Fiction vs. Nonfiction
pipe(sequence, fiction_categories)
```

```python
# Running the zero-shot classifier on the selected description and extracting the index of the highest-scoring label
max_index = np.argmax(pipe(sequence, fiction_categories)["scores"])

# Retrieving the label corresponding to the highest-scoring prediction
max_label = pipe(sequence, fiction_categories)["labels"][max_index]

# Displaying the predicted category label
max_label
```

```python
# Defining a helper function to generate zero-shot category predictions for a given text sequence
def generate_predictions(sequence, categories):
    """
    Generating a predicted category label for a given text sequence using zero-shot classification.

    This function performs the following steps:
    1. Running the zero-shot classifier on the input text to evaluate how well it matches each target category.
    2. Extracting the prediction scores returned by the model for each candidate label.
    3. Identifying the label with the highest score, representing the most likely category.
    4. Returning the predicted category as a single string.

    Parameters
    ----------
    sequence : str
        The input text (e.g., a book description) that we want to classify.

    categories : list
        A list of category labels that the model should choose from during zero-shot classification.

    Returns
    -------
    str
        The category label with the highest predicted probability.
    """
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
# Importing tqdm to display a progress bar during the prediction loop
from tqdm import tqdm

# Creating empty lists to store the true labels and the predicted labels
actual_cats = []
predicted_cats = []

# Looping through the first 300 Fiction book descriptions and generating predictions
for i in tqdm(range(0, 300)):
    # Selecting the i-th description from the books labeled as Fiction
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop = True)[i]

    # Appending the model's predicted category to the predictions list
    predicted_cats += [generate_predictions(sequence, fiction_categories)]

    # Appending the actual category label for evaluation
    actual_cats += ["Fiction"]
```

```python
# Looping through the first 300 Nonfiction book descriptions and generating predictions
for i in tqdm(range(0, 300)):
    # Selecting the i-th description from the books labeled as Nonfiction
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]

    # Appending the model's predicted category label to the predictions list
    predicted_cats += [generate_predictions(sequence, fiction_categories)]

    # Appending the true label ('Nonfiction') to the actual categories list
    actual_cats += ["Nonfiction"]
```

```python
# Creating a DataFrame to store actual and predicted category labels for evaluation
predictions_df = pd.DataFrame({
    "actual_categories": actual_cats,
    "predicted_categories": predicted_cats
})
```

```python
# Displaying the DataFrame containing the actual and predicted categories
predictions_df
```

```python
# Creating a new column that marks whether each prediction matches the actual category
# Assigning 1 for a correct prediction and 0 for an incorrect prediction
predictions_df["correct_prediction"] = (
    np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"], 1, 0)
)
```

```python
# Calculating overall prediction accuracy by dividing the number of correct predictions by the total number of samples
predictions_df["correct_prediction"].sum() / len(predictions_df)
```

```python
# Initializing an empty list to store the ISBN numbers of books with missing category labels
isbns = []

# Initializing an empty list to store the predicted categories for books without simplified labels
predicted_cats = []

# Selecting all books where the simplified category is missing
# Keeping only ISBN and description, and resetting the index for clean iteration
missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13", "description"]].reset_index(drop = True)
```

```python
# Looping through each book that is missing a simplified category label
for i in tqdm(range(0, len(missing_cats))):

    # Selecting the description text for the current book
    sequence = missing_cats["description"][i]

    # Generating a predicted category label using the zero-shot classifier
    predicted_cats += [generate_predictions(sequence, fiction_categories)]

    # Storing the corresponding ISBN number for tracking and assignment
    isbns += [missing_cats["isbn13"][i]]
```

```python
# Creating a DataFrame that pairs each ISBN with its predicted category label for books missing simplified categories
missing_predicted_df = pd.DataFrame({
    "isbn13": isbns,
    "predicted_categories": predicted_cats
})
```

```python
# Displaying the DataFrame containing ISBNs and predicted categories for books missing labels
missing_predicted_df
```

```python
# Merging the original books DataFrame with the DataFrame containing predicted categories
# Using a left join to keep all original rows while adding predictions where available
books = pd.merge(books, missing_predicted_df, on="isbn13", how="left")

# Filling missing simplified categories:
# If simple_categories is NA, replace it with the predicted category
# Otherwise, keep the existing simple_categories value
books["simple_categories"] = np.where(
    books["simple_categories"].isna(),
    books["predicted_categories"],
    books["simple_categories"]
)

# Dropping the temporary predicted_categories column now that missing categories have been filled
books = books.drop(columns=["predicted_categories"])
```

```python
# confusion matrix and detailed accuracy analysis
# importing metrics for evaluation
from sklearn.metrics import confusion_matrix, classification_report

print("\n" + "=" * 70)
print(" " * 20 + "Classification performance analysis")
print("=" * 70)

# create confusion matrix using predicted vs actual labels
cm = confusion_matrix(
    predictions_df["actual_categories"],
    predictions_df["predicted_categories"],
    labels = ["Fiction", "Nonfiction"],
)

# compute summary metrics
accuracy = predictions_df["correct_prediction"].sum() / len(predictions_df)

fiction_correct = predictions_df[
    (predictions_df["actual_categories"] == "Fiction")
    & (predictions_df["correct_prediction"] == 1)
].shape[0]

nonfiction_correct = predictions_df[
    (predictions_df["actual_categories"] == "Nonfiction")
    & (predictions_df["correct_prediction"] == 1)
].shape[0]

fiction_accuracy = fiction_correct / 300
nonfiction_accuracy = nonfiction_correct / 300

print("\nOverall metrics:")
print(f"  Total samples tested: {len(predictions_df):,}")
print(f"  Correct predictions: {predictions_df['correct_prediction'].sum():,}")
print(
    f"  Incorrect predictions: "
    f"{len(predictions_df) - predictions_df['correct_prediction'].sum():,}"
)
print(f"  Overall accuracy: {accuracy:.1%}")

print("\nCategory-specific accuracy:")
print(f"  Fiction accuracy: {fiction_accuracy:.1%} ({fiction_correct}/300)")
print(f"  Nonfiction accuracy: {nonfiction_accuracy:.1%} ({nonfiction_correct}/300)")

print("\nError analysis:")
print(f"  Fiction misclassified as nonfiction: {cm[0][1]}")
print(f"  Nonfiction misclassified as fiction: {cm[1][0]}")

# detailed classification report showing precision, recall and f1-score
print("\nDetailed classification report:")
print(
    classification_report(
        predictions_df["actual_categories"],
        predictions_df["predicted_categories"],
        target_names = ["Fiction", "Nonfiction"],
    )
)

# confusion matrix visualization
plt.figure(figsize = (8, 6))

sns.heatmap(
    cm,
    annot = True,
    fmt = "d",
    cmap = "Blues",
    xticklabels = ["Fiction", "Nonfiction"],
    yticklabels = ["Fiction", "Nonfiction"],
    cbar_kws = {"label": "Number of books"},
)

plt.xlabel("Predicted category", fontsize = 12)
plt.ylabel("Actual category", fontsize = 12)
plt.title(
    "Confusion matrix: zero-shot classification results\n(600 book evaluation set)",
    fontsize = 13,
)

# add percentage values on each matrix cell
for i in range(2):
    for j in range(2):
        percentage = cm[i][j] / 300 * 100
        plt.text(
            j + 0.5,
            i + 0.7,
            f"({percentage:.1f}%)",
            ha = "center",
            va = "center",
            fontsize = 10,
            color = "red",
        )

plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi = 300, bbox_inches = "tight")
plt.show()

print("\nConfusion matrix saved as 'confusion_matrix.png'")
print("=" * 70)
```

```python
# Category distribution analysis (before and after)
print("=" * 70)
print(" " * 20 + "Category distribution analysis")
print("=" * 70)

# Count categories before filling
books_before_fill = pd.read_csv(
    "/content/drive/MyDrive/Colab Notebooks/NLP/NLP Project/books_cleaned.csv"
)
books_before_fill["simple_categories"] = books_before_fill["categories"].map(
    category_mapping
)

# Statistics before filling
total_books = len(books_before_fill)
labeled_before = books_before_fill["simple_categories"].notna().sum()
missing_before = books_before_fill["simple_categories"].isna().sum()

# Statistics after filling
labeled_after = books["simple_categories"].notna().sum()
missing_after = books["simple_categories"].isna().sum()

print("\nBefore llm classification:")
print(f"  Total books: {total_books:,}")
print(
    f"  Labeled (manual mapping): "
    f"{labeled_before:,} ({labeled_before / total_books * 100:.1f}%)"
)
print(
    f"  Missing labels: "
    f"{missing_before:,} ({missing_before / total_books * 100:.1f}%)"
)

print("\nAfter llm classification:")
print(f"  Total books: {total_books:,}")
print(
    f"  Labeled (manual + llm): "
    f"{labeled_after:,} ({labeled_after / total_books * 100:.1f}%)"
)
print(
    f"  Missing labels: "
    f"{missing_after:,} ({missing_after / total_books * 100:.1f}%)"
)
print(f"  Books classified by llm: {labeled_after - labeled_before:,}")

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 6))

# Before filling
before_counts = books_before_fill["simple_categories"].value_counts()
before_counts["Missing"] = missing_before

colors_before = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]

ax1.bar(
    range(len(before_counts)),
    before_counts.values,
    color = colors_before[: len(before_counts)],
)

ax1.set_xticks(range(len(before_counts)))
ax1.set_xticklabels(before_counts.index, rotation = 45, ha = "right")
ax1.set_ylabel("Number of books", fontsize = 12)
ax1.set_title(
    "Category distribution before llm\n(manual mapping only)",
    fontsize = 13,
)
ax1.grid(True, alpha = 0.3, axis = "y")

# Add value labels on bars
for i, v in enumerate(before_counts.values):
    ax1.text(
        i,
        v + 50,
        str(int(v)),
        ha = "center",
        va = "bottom",
        fontweight = "bold",
    )

# After filling
after_counts = books["simple_categories"].value_counts()
if missing_after > 0:
    after_counts["Missing"] = missing_after

colors_after = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]

ax2.bar(
    range(len(after_counts)),
    after_counts.values,
    color = colors_after[: len(after_counts)],
)

ax2.set_xticks(range(len(after_counts)))
ax2.set_xticklabels(after_counts.index, rotation = 45, ha = "right")
ax2.set_ylabel("Number of books", fontsize = 12)
ax2.set_title(
    "Category distribution after llm\n(manual + zero-shot)",
    fontsize = 13,
)
ax2.grid(True, alpha = 0.3, axis = "y")

# Add value labels on bars
for i, v in enumerate(after_counts.values):
    ax2.text(
        i,
        v + 50,
        str(int(v)),
        ha = "center",
        va = "bottom",
        fontweight = "bold",
    )

plt.tight_layout()
plt.savefig("category_distribution_comparison.png", dpi = 300, bbox_inches = "tight")
plt.show()

print("\nVisualization saved as 'category_distribution_comparison.png'")
print("=" * 70)
```
